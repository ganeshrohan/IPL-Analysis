```python
#Importing required packages 

import pandas as pd 
import matplotlib
from matplotlib import pyplot as plt
```

```python
#Loading dataset using pandas 
ipl = pd.read_csv("matches.csv")
#Looking at first 5 records 
ipl.head()
```

```python
#No. of rows and columns in datasets
ipl.shape
```

```python
ipl.info()
```

```python
#Getting the top 5 players frequency for "man of the match"
ipl['player_of_match'].value_counts()[0:5]
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(ipl["player_of_match"].value_counts()[0:5].keys()),list(ipl["player_of_match"].value_counts()[0:5]), color='b')
plt.show()
```

```python
#Looking at the result
ipl["result"].value_counts()
```

```python
#Finding the Nunmber of toss wins for each team
ipl["toss_winner"].value_counts()
```

```python
#Getting the teams that won matches by batting first
won_batting_first = ipl[ipl["win_by_runs"]!=0]
won_batting_first.head()
```

```python
#plotting histogram
plt.hist(won_batting_first["win_by_runs"],bins=15)
plt.title("Distribution by runs")
plt.xlabel("Runs")
plt.ylabel("No. of matches")
plt.show()
```

```python
#Teams that won most matches by batting first
won_batting_first["winner"].value_counts()
```

```python
#Plotting top 5 teams 

plt.figure(figsize=(15,8))
plt.bar(list(won_batting_first["winner"].value_counts()[0:5].keys()),list(won_batting_first["winner"].value_counts()[0:5]),color=['blue','yellow','orange','purple','red'])
plt.ylabel("No. of matches won")
plt.xlabel("Teams")
plt.show()
```

```python
#Pie chart to know teams that win by batting first
plt.figure(figsize=(20,20))
plt.pie(list(won_batting_first["winner"].value_counts()),labels= list(won_batting_first["winner"].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#Teams won batting second
won_batting_second = ipl[ipl["win_by_wickets"]!=0]
won_batting_second.head()
```

```python
#No. of time teams won the match by batting first
won_batting_second["winner"].value_counts()
```

```python
#Histogram
plt.hist(won_batting_second["win_by_wickets"],bins=20)
plt.xlabel("No. of wickets remaining")
plt.ylabel("No. of matches")
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.bar(list(won_batting_second["winner"].value_counts()[0:5].keys()),list(won_batting_second["winner"].value_counts()[0:5]),color=['blue','yellow','red','purple','orange'])
plt.ylabel("No. of matches won")
plt.xlabel("Teams")
plt.show()
```

```python
#Pie chart to know teams that won by batting second
plt.figure(figsize=(20,20))
plt.pie(list(won_batting_second["winner"].value_counts()),labels= list(won_batting_second["winner"].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
